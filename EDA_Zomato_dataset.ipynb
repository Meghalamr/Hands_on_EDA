```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
value=final_df.Country.value_counts().values
value
```

```python
#Pie Chart
plt.pie(value,labels=country_names)
```

#The above pie chart is not so much good. Lets plot the pie chart for the top 3 countries that uses zomato

```python
plt.pie(value[:3],labels=country_names[:3])
```

# need to represent the pie chart with percentage

```python
plt.pie(value[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation = Zomato maximum records taken place in India, secondly in United states with  and thirdly in United kingdom among the world.

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

# I observed that 4.5 - 4.9 ------> Excellent
4.0-4.5 ------> Very Good

3.5-3.9 ------> Good

2.5-3.4 ------> Average

1.8-2.4 ------> Poor

0 ------> Not rated

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green'])
```

# Observation

1.Not Rated count is very high

2.Maximum rating count is 3.1 and 3.2

# for categorical feature the plot used
Count plot

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green'])
```
